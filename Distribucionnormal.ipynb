```python
import random 
import matplotlib.pyplot as plt
media=5 #aqui se pone la media o mu 
desviacion=10 #aqui se pone la desviacion o sigma 

#doce uniforme 
numeros=[]
for i in range(10000):
  b=0
  for i in range(12):
    a=random.random()
    b=a+b
  x=media+(desviacion*(b-6))
  numeros.append(x)

#funcion de python 
nums = []

for i in range(10000):
    temp = random.gauss(media, desviacion)
    nums.append(temp)

plt.title("Función de Python")
plt.hist(nums, bins = 100)
plt.show()

plt.title("Doce uniformes")
plt.hist(numeros, bins =100 )
plt.show()
```
